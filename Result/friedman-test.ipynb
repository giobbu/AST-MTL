```r
library(tsutils)

x <- as.matrix(read.csv("../input/loss-mtl/df_FLOW_RMSE.csv", sep=",",header = TRUE, row.names = 1) )

dimnames(x) <- NULL


colnames(x) <- c('HA',
                 'AST-MTL',
                   'STL_GCN_GRU',
                   'amtx_STL',
                   'MTL_GRU',
                   'STL_GRU', 'STL_LSTM')



# jpeg(file="FLOW-RMSE.jpeg", width=600)
nemenyi(x,conf.level=0.95,plottype="vline")
# dev.off()
```
